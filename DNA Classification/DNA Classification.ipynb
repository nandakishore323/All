```python
import numpy as np
import pandas as pd
import sklearn
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/molecular-biology/promoter-gene-sequences/promoters.data'
names = ['Class','id','Sequence']
df = pd.read_csv(url, names=names)
```

```python
df.head()
```

```python
classes = df.loc[:,'Class']
```

```python
sequences = list(df.loc[:, 'Sequence'])
dataset = {}

for i,seq in enumerate(sequences):
    
    nucleotides = list(seq)
    nucleotides = [x for x in nucleotides if x!='\t']
    
    nucleotides.append(classes[i])
    
    dataset[i] = nucleotides
    
```

```python
df = pd.DataFrame(dataset)
```

```python
df.head()
```

```python
df = df.transpose()
```

```python
df.head()
```

```python
df.rename(columns = {57: 'Class'}, inplace = True)
df.head()
```

```python
df.describe()
```

```python
num_df = pd.get_dummies(df)
num_df.head()
```

```python
df = num_df.drop(columns = ['Class_-'])
df.rename(columns = {'Class_+' : 'Class'},inplace = True)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.metrics import classification_report , accuracy_score
```

```python
scoring = 'accuracy'

from sklearn.model_selection import train_test_split,KFold,cross_val_score

X = np.array(df.drop(['Class'], 1))
y = np.array(df['Class'])

X_train, X_test , y_train , y_test = train_test_split(X,y)
```

```python
model = KNeighborsClassifier(n_neighbors = 2)
kfold = KFold(n_splits = 10)
cv_results = cross_val_score(model, X_train , y_train, cv = kfold , scoring = scoring)
print(cv_results.mean())
```

```python
model.fit(X_train,y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier(max_depth = 7)
kfold = KFold(n_splits = 10)
cv_results = cross_val_score(model, X_train , y_train, cv = kfold , scoring = scoring)
print(cv_results.mean())
```

```python
model.fit(X_train,y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = RandomForestClassifier(n_estimators = 1000,max_depth=7, random_state=0)
kfold = KFold(n_splits = 10)
cv_results = cross_val_score(model, X_train , y_train, cv = kfold , scoring = scoring)
print(cv_results.mean())
```

```python
model.fit(X_train,y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```

```python
model = SVC(kernel = 'linear')
kfold = KFold(n_splits = 10)
cv_results = cross_val_score(model, X_train , y_train, cv = kfold , scoring = scoring)
print(cv_results.mean())
```

```python
model.fit(X_train,y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```

```python
model = SVC(kernel = 'sigmoid', C = 1000.0)
kfold = KFold(n_splits = 10)
cv_results = cross_val_score(model, X_train , y_train, cv = kfold , scoring = scoring)
print(cv_results.mean())
```

```python
model.fit(X_train,y_train)
pred = model.predict(X_test)
print(accuracy_score(y_test,pred))
print(classification_report(y_test,pred))
```
